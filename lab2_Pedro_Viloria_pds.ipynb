## **2.3 (submuestreo)**

# **3 Frecuencia de Nyquist y Aliasing**

## **3.1**

```python
import numpy as np
import matplotlib.pyplot as plt

# Frecuencia de la primera señal sinusoidal en Hz
f1 = 1

# Frecuencia de la segunda señal sinusoidal en Hz
f2 = 1

# Frecuencia de muestreo de la primera señal en Hz
fs1 = 20

# Frecuencia de muestreo de la segunda señal en Hz
fs2 = 2*fs1

# Duración de la señal en segundos
duration = 5

# Número de puntos de la señal con frecuencia de muestreo más baja
n = int(fs1 * duration)

# Generar la primera señal sinusoidal
t1 = np.linspace(0, duration, n, endpoint=False)
x1 = np.sin(2 * np.pi * f1 * t1)

# Generar la segunda señal sinusoidal
t2 = np.linspace(0, duration, int(fs2 * duration), endpoint=False)
x2 = np.sin(2 * np.pi * f2 * t2)

# Interpolar la segunda señal sinusoidal a la frecuencia de muestreo de la primera señal
x2_interp = np.interp(t1, t2, x2)

# Graficar ambas señales sinusoidales en la misma gráfica
plt.plot(t1, x1, label='Señal 1', linewidth=3.4,color="blue",linestyle= "--")
plt.plot(t1, x2_interp, label='Señal 2',linewidth=2,color="red")
plt.title("SEÑALES SINUSOIDALES")
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.legend()
plt.grid(True)
plt.show()
```

A simple vista es dificil encontrar diferencias, lo que es posible observar es que hay una correspondencia entre las señales, puesto que se acoplan de la misma manera. Lo que para efectos de señales sería una gran ayuda al momento de escoger alguna para enviar información.

# **4. Cuantificación**

## **4.1**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile

def fxquant(s, bit):
    # s: señal de entrada, debe estar normalizada entre -1 y 1
    # bit: bits de cuantizacion
    Plus1 = np.power(2, (bit - 1))
    X = s * Plus1
    X = np.round(X)
    X = np.minimum(Plus1 - 1.0, X)
    X = np.maximum(-1.0 * Plus1, X)
    X = X / Plus1
    return X

# cargar la señal de audio
fs, audio = wavfile.read('8.wav')

# normalizar la señal entre -1 y 1
audio_norm = audio / np.max(np.abs(audio))

# re-cuantificar la señal a 2 bits
audio_2bits = fxquant(audio_norm, 2)

# re-cuantificar la señal a 4 bits
audio_4bits = fxquant(audio_norm, 4)

# re-cuantificar la señal a 12 bits
audio_12bits = fxquant(audio_norm, 12)

# guardar las señales re-cuantificadas en archivos .wav
wavfile.write('audio_2bits.wav', fs, np.int16(audio_2bits * 32767))
wavfile.write('audio_4bits.wav', fs, np.int16(audio_4bits * 32767))
wavfile.write('audio_12bits.wav', fs, np.int16(audio_12bits * 32767))

# graficar las señales re-cuantificadas y la señal original
plt.figure(figsize=(14,6))
plt.subplot(2,2,1)
plt.plot(audio_norm)
plt.title('Señal Original')
plt.grid(True)
plt.subplot(2,2,2)
plt.plot(audio_2bits)
plt.title('Señal Re-cuantificada a 2 bits')
plt.grid(True)
plt.subplot(2,2,3)
plt.plot(audio_4bits)
plt.title('Señal Re-cuantificada a 4 bits')
plt.grid(True)
plt.subplot(2,2,4)
plt.plot(audio_12bits)
plt.title('Señal Re-cuantificada a 12 bits')
plt.grid(True)
plt.show()


# reproducir el archivo de audio re-cuantificado a 2 bits
Audio('audio_2bits.wav')

# reproducir el archivo de audio re-cuantificado a 4 bits
Audio('audio_4bits.wav')

# reproducir el archivo de audio re-cuantificado a 12 bits
Audio('audio_12bits.wav')

```

#5.1 **Generación de señales analógica y digital**

```python
import scipy.io.wavfile as wavfile
import matplotlib.pyplot as plt
import numpy as np

# Cargar la señal de audio desde un archivo utilizando scipy
sr, audio = wavfile.read('8.wav')

# Normalizar la señal de audio en amplitud
audio_norm = audio / np.abs(audio).max()

# Re-escalar la señal de audio en un factor de 460
audio_rescaled = audio_norm * 460

# Calcular la señal digital aproximando la señal analógica
audio_digital = np.round(audio_rescaled)

# Agregar 100 muestras intermedias a la señal analógica
time_interp = np.linspace(0, len(audio_rescaled) / sr, num=len(audio_rescaled) * 100)
audio_rescaled_interp = np.interp(time_interp, np.arange(len(audio_rescaled)) / sr, audio_rescaled)

# Agregar 100 muestras intermedias a la señal digital
audio_digital_interp = np.interp(time_interp, np.arange(len(audio_digital)) / sr, audio_digital)

# Graficar la señal analógica y digital con muestras intermedias
fig, axs = plt.subplots(2, 1, figsize=(10, 6))
axs[0].plot(time_interp, audio_rescaled_interp)

axs[0].set_xlabel('Tiempo (s)')
axs[0].set_ylabel('Amplitud')

axs[0].set_title('Señal analógica')

axs[1].plot(time_interp, audio_digital_interp)
axs[1].set_xlabel('Tiempo (s)')
axs[1].set_ylabel('Amplitud')

axs[1].set_title('Señal digital')
plt.grid(True)
plt.tight_layout()
plt.show()
```

En las señales graficadas por ese código se puede ver una diferencia entre la señal analógica y la señal digital. La señal analógica, que es la señal re-escalada, tiene una forma de onda suave y continua, con transiciones suaves entre valores. Por otro lado, la señal digital, que es la señal re-escalada y luego redondeada a los valores enteros más cercanos, tiene una forma de onda escalonada, con transiciones bruscas entre valores discretos.

La diferencia se puede apreciar en la figura, donde la señal analógica se representa con una línea continua y la señal digital se representa con una línea punteada. La señal analógica puede tener una mayor fidelidad en la representación de la señal original, ya que no pierde información en la discretización, pero la señal digital es más fácil de procesar y almacenar en un formato digital.

# **5.2 Transmisión**

```python
import numpy as np
import matplotlib.pyplot as plt

# Función para atenuar la señal, agregar ruido y amplificarla nuevamente
def repeater(x, noise_amplitude, attenuation):
    x = x * attenuation
    noise = np.random.uniform(-noise_amplitude, noise_amplitude, len(x))
    x = x + noise
    x = x / attenuation
    return x

# Función para simular la transmisión analógica
def analog_tx(x, num_repeaters, noise_amplitude, attenuation):
    for n in range(0, num_repeaters):
        x = repeater(x, noise_amplitude, attenuation)
    return x

# Función para simular la transmisión digital
def digital_tx(x, num_repeaters, noise_amplitude, attenuation):
    for n in range(0, num_repeaters):
        x = np.round(repeater(x, noise_amplitude, attenuation))
    return x

# Función para calcular la SNR
def SNR(noisy, original):
    err = np.var(original - noisy)
    sig = np.var(original)
    return 10 * np.log10(sig/err)

# Parámetros de la simulación
num_repeaters = 150
noise_amplitude = 0.15
attenuation = 0.35

# Generación de señal de prueba
fs = 1000
t = np.arange(0, 1, 1/fs)
f1 = 10
f2 = 200
signal = np.sin(2 * np.pi * f1 * t) + 0.2 * np.sin(2 * np.pi * f2 * t)

# Transmisión de la señal analógica
analog_signal = analog_tx(signal, num_repeaters, noise_amplitude, attenuation)
analog_snr = SNR(analog_signal, signal)

# Transmisión de la señal digital
digital_signal = digital_tx(signal, num_repeaters, noise_amplitude, attenuation)
digital_snr = SNR(digital_signal, signal)

print(f"SNR señal analógica: {analog_snr:.2f} dB")
print(f"SNR señal digital: {digital_snr:.2f} dB")

# Barrido del número de repetidores
num_rep_values = np.arange(10, 101, 10)
analog_snrs = []
digital_snrs = []
for num_rep in num_rep_values:
    analog_signal = analog_tx(signal, num_rep, noise_amplitude, attenuation)
    analog_snr = SNR(analog_signal, signal)
    analog_snrs.append(analog_snr)

    digital_signal = digital_tx(signal, num_rep, noise_amplitude, attenuation)
    digital_snr = SNR(digital_signal, signal)
    digital_snrs.append(digital_snr)

# Gráfico de la SNR en función del número de repetidores
plt.plot(num_rep_values, analog_snrs, label="Señal analógica")
plt.plot(num_rep_values, digital_snrs, label="Señal digital")
plt.legend()
plt.xlabel("Número de repetidores")
plt.ylabel("SNR (dB)")
plt.show()
```
